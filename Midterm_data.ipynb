```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Imported Delta passenger data into a DataFrame
delta_df = pd.read_excel(r'Delta_Airlines_Passengers.xls')
print(delta_df)
# pd.set_option('display.max_rows', 500)
```

This dataset has 5 variables including 2 categorical *(years and months)* and 3 numerical *(domestic, international and total)*

- DOMESTIC:  Number of passengers who traveled within the US on Delta flights. 
- INTERNATIONAL: Number of passengers who traveled outside the US on Delta flights.
- TOTAL: Number of passengers who traveled domestically and internationally on Delta flights.
- YEAR: 2002 - 2022
- MONTHS: January - December

There are 243 entries since the data is collected on a monthly basis from October 2002 to December 2022. We can see immediately there are some data missing for the last three months of 2022. This is because the most recent three months of international data by airport and carrier are withheld due to confidentiality agreements. However, the number of missing data are small enough that it should not affect our analysis. 

## Description of Variables
### Domestic

```python
# Domestic passenger numbers in millions
domestic = delta_df['DOMESTIC']/1000000
# Summary statistics for domestic passengers
domestic.describe()
```

The domestic data shows us the number of passengers who traveled within the US on Delta flights every month from 2002 to 2022. The average number of domestic travelers in a month is 7.65 million, with a standard deviation of 2.41 million. Domestic travel reached a maximum of 12.97 million passengers in July 2019 and a minimum of 0.45 million in April 2020. 

```python
# Domestic histogram
x = [domestic]
plt.hist(x, bins=10)
plt.title('Histogram - Number of domestic passengers in the US 2002-2022')
plt.xlabel('domestic passengers in millions')
plt.ylabel('frequency')
plt.show()

# Domestic boxplot
sns.boxplot(y=domestic)
plt.ylabel('domestic passengers in millions')
plt.title('Boxplot - Number of domestic passengers in the US 2002-2022')
plt.show()
```

Based on the histogram and boxplot the distribution of domestic passengers is unimodal, unsymmetical and a little left skewed. However, the mean (7.65 millions) and median (7.66 million) are very close which indicates a normal distribution and there are no outliers.

### International

```python
# International passenger numbers in millions
international = delta_df['INTERNATIONAL']/1000000
# Summary statistics for international passengers
international.describe()
```

The international data represents the number of passengers who traveled outside the US on Delta flights every month from 2002 to 2022. The average number of international travelers in a month is 1.34 million, with a standard deviation of 0.63 million. International travel reached a maximum of 2.63 million passengers also in July 2019 and a minimum of about 0.03 million in May 2020.

```python
# International histogram
x = [international]
plt.hist(x, bins=14)
plt.title('Histogram - Number of international passengers in the US 2002-2022')
plt.xlabel('international passengers in millions')
plt.ylabel('frequency')
plt.show()

# Internatinal boxplot
sns.boxplot(y=international)
plt.ylabel('international passengers in millions')
plt.title('Boxplot - Number of international passengers in the US 2002-2022')
plt.show()
```

Looking at the histogram and boxplot for international passengers, the distribution appears to be unsymmetrical, bimodal and slightly left skewed. There are no outliers and with the mean (1.35 million) and median (1.41) being quite close we can assume a somewhate normal distribution.

### Total

```python
# Total passenger numbers in millions
total = delta_df['TOTAL']/1000000
# Summary statistics for total passengers
total.describe()
```

The total data represents the number of passengers who traveled on all Delta flights every month from 2002 to 2022 which includes domestic and international travel. The average number of total travelers *(domestic + international)* in a month is 8.96 million, with a standard deviation of 2.95 million. Total travel reached a maximum of 15.60 million passengers in July 2019 and a minimum of 0.49 million in April 2020.

```python
# Total histogram
x = [total]
plt.hist(x, bins=10)
plt.title('Histogram - Number of total passengers in the US 2002-2022')
plt.xlabel('total passengers in millions')
plt.ylabel('frequency')
plt.show()

# Total boxplot
sns.boxplot(y=total)
plt.ylabel('total passengers in millions')
plt.title('Boxplot - Number of total passengers in the US 2002-2022')
plt.show()
```

The distribution of the total number of passengers on Delta flights appears to be left skewed, unsymmetrical and bimodal. Most months in the 20 year period had around 6 million passengers which is our mode. The boxplot does not display any outliers and the mean (8.96 million) and median (9.05 million) are very close so this is also a normal distribution.

## Relationships between variables
### Domestic vs International

Is there a correlation between the number domestic and international passengers?
We expect that as the number of domestic passengers increases so do the number of international passengaers. To examine this, we look at the scatterplots for these variables and the regression line.

```python
years = delta_df['Year']
months = delta_df['Month']
# creating scatterplot comparing domestic and international passengers
domestic_international_scatter = sns.regplot(x=domestic, y=international)
domestic_international_scatter.set(xlabel='domestic passengers', ylabel='international passengers', 
                                   title='Scatter plot of domestic VS international travelers in millions')
```

The scatterplot aboves shows a strong, positive and linear relationship between the number of domestic and international passengers. International numbers do increase as domestic numbers increase.

```python
# Correlations
corr = delta_df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation matrix')
plt.show()
```

Looking at the correlations matrix for domestic against international, we get a coefficient of correlation of 0.85 which reinforces that there is a strong relationship between them.

```python
# Comparing domestic and international scatterplots
plt.scatter(years, domestic, label = "domestic", color='r')
plt.scatter(years, international, label = "international", color='g')
plt.xlabel('years') 
plt.ylabel('passengers in millions')
plt.title('Scatterplot comparison of domestic and international passengers')
plt.legend()
```

Each dot in the scatterplot above represents one month with each vertical cluster of those dots representing one year in their respective colors. We can see with each cluster that passenger numbers for both domestic and international options fluctuates within a particular year because the highest travel season is during the summer and declines a little in the winter. There are two majors periods where travel has signicantly decreased in 2006-2009 and 2020-2021. The reason for the 2006-2009 decline was because of The Great Recession when global economies were declining due to the housing crisis, banks collapsing and the stock market crashing. During this period of economic uncertaintly, many people were hesitant to travel due to financial concerns. The 2020-2021 period was when the world went into lockdown as the pandemic halted almost all travel. However, after much of the world had been inoculated against the virus air travel started to increase again.
