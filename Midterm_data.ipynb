In order to understand the impact of global events on the airline industry, we are analysing data on the number of passengers who traveled with Delta Airlines collected by the US Bureau of Transportation Statistics that spans from the year 2002 to 2022. This dataset includes domestic travel, international travel, years, months and the total number of travellers of the entire sample. As Delta Airlines is the largest carrier in the world with many domestic and internatiional routes, we believe they are a good representation of the airline industry. Airlines also play a key role in the economic development of a country and therefore Delta's performance can indicate the health of the US economy. This data was collected on a monthly basis.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Imported Delta passenger data into a DataFrame
delta_df = pd.read_excel(r'Delta_Airlines_Passengers.xls')
print(delta_df)
```

```python
# Summary statistics on single variables
summary = delta_df.describe()
print(summary)
```

```python
domestic = delta_df['DOMESTIC']
international = delta_df['INTERNATIONAL']
domestic_and_international = delta_df['TOTAL']

x = [domestic]
plt.hist(x, bins=10)
plt.show()
```

There are 5 variables in this dataset with two being categoricals (months and years) and three numericals (domestic, international, and total). 

```python
# Bivariate analysis
corr = delta_df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation matrix')
plt.show()
```

```python
# Exploratory data analysis
sns.pairplot(delta_df)
plt.show()
```
