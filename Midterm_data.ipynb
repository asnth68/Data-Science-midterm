```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Imported Delta passenger data into a DataFrame
delta_df = pd.read_excel(r'Delta_Airlines_Passengers.xls')
print(delta_df)
# pd.set_option('display.max_rows', 500)
```

This dataset has 5 variables including 2 categorical *(years and months)* and 3 numerical *(domestic, international and total)*

- DOMESTIC:  Number of passengers who traveled within the US on Delta flights. 
- INTERNATIONAL: Number of passengers who traveled outside the US on Delta flights.
- TOTAL: Number of passengers who traveled domestically and internationally on Delta flights.
- YEAR: 2002 - 2022
- MONTHS: January - December

There are 243 entries since the data is collected on a monthly basis from October 2002 to December 2022. We can see immediately there are some data missing for the last three months of 2022. This is because the most recent three months of international data by airport and carrier are withheld due to confidentiality agreements. However, the number of missing data are small enough that it should not affect our analysis. 

## Description of Variables
### Domestic

```python
# Domestic passenger numbers in millions
domestic = delta_df['DOMESTIC']/1000000
# Summary statistics for domestic passengers
domestic.describe()
```

The domestic data shows us the number of passengers who traveled within the US on Delta flights every month from 2002 to 2022. The average number of domestic travelers in a month is 7.65 million, with a standard deviation of 2.41 million. Domestic travel reached a maximum of 12.97 million passengers in July 2019 and a minimum of 0.45 million in April 2020. 

```python
# Domestic histogram
x = [domestic]
plt.hist(x, bins=10)
plt.title('Histogram - Number of domestic passengers in the US 2002-2022')
plt.xlabel('domestic passengers in millions')
plt.ylabel('frequency')
plt.show()

# Domestic boxplot
sns.boxplot(y=domestic)
plt.ylabel('domestic passengers in millions')
plt.title('Boxplot - Number of domestic passengers in the US 2002-2022')
plt.show()
```

Based on the histogram and boxplot the distribution of domestic passengers is unimodal, unsymmetical and a little left skewed. However, the mean (7.65 millions) and median (7.66 million) are very close which indicates a normal distribution and there are no outliers.

```python
# International passenger numbers in millions
international = delta_df['INTERNATIONAL']/1000000
# Summary statistics for international passengers
international.describe()
```

The average number of international travelers in a month is 1.34 million, with a standard deviation of 0.63 million.
